```python
from PIL import Image, ImageFilter
import numpy as np
from scipy import ndimage

def detect_edges(image_path):
    # Завантаження зображення
    image = Image.open(image_path).convert("L")  # Конвертація у відтінки сірого
    
    # Перетворення в numpy-масив
    image_array = np.array(image)
    
    # Застосування фільтра для виділення контурів
    edges = ndimage.sobel(image_array)
    
    # Перетворення назад у зображення
    edge_image = Image.fromarray(edges)
    edge_image.show()
    
    return edge_image

# Виклик функції для тесту
detect_edges('roxsana.png')
```
