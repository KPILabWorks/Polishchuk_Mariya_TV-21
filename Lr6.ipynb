```python
import pandas as pd
import matplotlib.pyplot as plt

# Завантаження даних з CSV файлу
try:
    df = pd.read_csv("Raw Data.csv")
except FileNotFoundError:
    print("Помилка: Файл 'Raw Data.csv' не знайдено. Переконайтеся, що файл знаходиться в тому ж каталозі, що і скрипт.")
    exit()
except Exception as e:
    print(f"Помилка при завантаженні файлу: {e}")
    exit()

# Перевірка наявності необхідних стовпців
required_columns = ["Time (s)", "Magnetic Field x (µT)", "Magnetic Field y (µT)", "Magnetic Field z (µT)", "Absolute field (µT)"]
for col in required_columns:
    if col not in df.columns:
        print(f"Помилка: Стовпець '{col}' відсутній у файлі CSV.")
        exit()

# Створення графіків для кожної компоненти та абсолютного значення
plt.figure(figsize=(12, 8))

# Magnetic Field x
plt.subplot(2, 2, 1)
plt.plot(df["Time (s)"], df["Magnetic Field x (µT)"])
plt.xlabel("Час (с)")
plt.ylabel("Магнітне поле x (µT)")
plt.title("Магнітне поле x в залежності від часу")
plt.grid(True)

# Magnetic Field y
plt.subplot(2, 2, 2)
plt.plot(df["Time (s)"], df["Magnetic Field y (µT)"])
plt.xlabel("Час (с)")
plt.ylabel("Магнітне поле y (µT)")
plt.title("Магнітне поле y в залежності від часу")
plt.grid(True)

# Magnetic Field z
plt.subplot(2, 2, 3)
plt.plot(df["Time (s)"], df["Magnetic Field z (µT)"])
plt.xlabel("Час (с)")
plt.ylabel("Магнітне поле z (µT)")
plt.title("Магнітне поле z в залежності від часу")
plt.grid(True)

# Absolute Field
plt.subplot(2, 2, 4)
plt.plot(df["Time (s)"], df["Absolute field (µT)"])
plt.xlabel("Час (с)")
plt.ylabel("Абсолютне поле (µT)")
plt.title("Абсолютне поле в залежності від часу")
plt.grid(True)

plt.tight_layout()  # Запобігання перекриттю графіків
plt.savefig("magnetic_field_components_ukr.png")  # Збереження графіку в файл
plt.show() # Відображення графіку на екрані
```
