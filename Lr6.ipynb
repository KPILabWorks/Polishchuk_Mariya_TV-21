```python
import pandas as pd
import matplotlib.pyplot as plt

#    CSV 
try:
    df = pd.read_csv("Raw Data.csv")
except FileNotFoundError:
    print(":  'Raw Data.csv'  . ,       ,   .")
    exit()
except Exception as e:
    print(f"   : {e}")
    exit()

#    
required_columns = ["Time (s)", "Magnetic Field x (T)", "Magnetic Field y (T)", "Magnetic Field z (T)", "Absolute field (T)"]
for col in required_columns:
    if col not in df.columns:
        print(f":  '{col}'    CSV.")
        exit()

#        
plt.figure(figsize=(12, 8))

# Magnetic Field x
plt.subplot(2, 2, 1)
plt.plot(df["Time (s)"], df["Magnetic Field x (T)"])
plt.xlabel(" ()")
plt.ylabel("  x (T)")
plt.title("  x    ")
plt.grid(True)

# Magnetic Field y
plt.subplot(2, 2, 2)
plt.plot(df["Time (s)"], df["Magnetic Field y (T)"])
plt.xlabel(" ()")
plt.ylabel("  y (T)")
plt.title("  y    ")
plt.grid(True)

# Magnetic Field z
plt.subplot(2, 2, 3)
plt.plot(df["Time (s)"], df["Magnetic Field z (T)"])
plt.xlabel(" ()")
plt.ylabel("  z (T)")
plt.title("  z    ")
plt.grid(True)

# Absolute Field
plt.subplot(2, 2, 4)
plt.plot(df["Time (s)"], df["Absolute field (T)"])
plt.xlabel(" ()")
plt.ylabel("  (T)")
plt.title("     ")
plt.grid(True)

plt.tight_layout()  #   
plt.savefig("magnetic_field_components_ukr.png")  #    
plt.show() #    
```
